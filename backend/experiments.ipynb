```python
import json
import re
from agno.run.response import RunEvent, RunResponse

def parse_mixed_response(response, stream):
    """
    Processes Agno response, extracting JSON and separating additional text.

    Args:
        response: The response object from `agent.run()`
        stream (bool): Whether to process the response as a stream

    Returns:
        tuple: (parsed_json, additional_text)
    """
    full_text = ""

    if stream:
        collected_chunks = []
        for chunk in response:
            if isinstance(chunk, RunResponse) and isinstance(chunk.content, str):
                if chunk.event == RunEvent.run_response:
                    collected_chunks.append(chunk.content)
        
        full_text = "".join(collected_chunks)
    else:
        for chunk in response:
            if isinstance(chunk, RunResponse) and isinstance(chunk.content, str):
                if chunk.event == RunEvent.run_response:
                    full_text += chunk.content
    extracted_json = None
    additional_text = full_text
    json_match = re.search(r"\{.*\}", full_text, re.DOTALL)
    if json_match:
        json_str = json_match.group(0) 
        try:
            extracted_json = json.loads(json_str)
            additional_text = full_text.replace(json_str, "").strip()  
        except json.JSONDecodeError:
            pass 

    return extracted_json, additional_text
```
